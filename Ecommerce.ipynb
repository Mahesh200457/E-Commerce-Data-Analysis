```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default = "plotly_white"
import seaborn as sns
```

```python
df=pd.read_csv("superstore.csv",encoding="latin-1")
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
df["Order Date"]=pd.to_datetime(df["Order Date"])
df["Ship Date"]=pd.to_datetime(df["Ship Date"])
```

```python
df.head()
```

```python
df.info()
```

```python
df["order Month"]=df["Order Date"].dt.month
df["order Year"]=df["Order Date"].dt.year
df["order Day"]=df["Order Date"].dt.day
```

```python
df.head()
```

```python
df["order Month"]=pd.to_numeric(df["order Month"])
```

```python
month_label={1:"jan",2:"feb",3:"mar",4:"apr",5:"may",6:"jun",7:"jul",8:"aug",9:"sep",10:"oct",11:"nov",12:"dec"}
df["order Month"]=df["order Month"].map(month_label)
df.head()
```

```python
df["order Day"] = df["Order Date"].dt.day_name()
df.head()
```

**1. Monthly Sales Analysis**

```python
sales_by_month=df.groupby("order Month")["Sales"].sum().reset_index()


sales_by_month
```

```python
fig=px.line(sales_by_month,
            x="order Month",
            y="Sales",
            title="Monthly_Sales_Analysis")
fig.show()
```

```python
# Sort the DataFrame by the original numeric month order
month_order = list(month_label.values())
sales_by_month["order Month"] = pd.Categorical(sales_by_month["order Month"], categories=month_order, ordered=True)
sales_by_month = sales_by_month.sort_values("order Month")
```

```python
fig=px.line(sales_by_month,
            x="order Month",
            y="Sales",
            title="Monthly_Sales_Analysis")
fig.show()
```

**2. Category-Wise Sales Analysis**

ðŸ”¸ Analyze sales based on product categories.

ðŸ”¸ Find out which category had the highest and lowest sales.



```python
sales_by_category=df.groupby("Category")["Sales"].sum().reset_index()
```

```python
sales_by_category
```

```python
fig=px.pie(sales_by_category,
           values="Sales",
           names="Category",hole=0.5,
           color_discrete_sequence=px.colors.sequential.RdBu)
fig.update_traces(textposition="inside",textinfo="percent+label")

fig.update_layout(title_text="Sales By Category",title_font=dict(size=24),title_x=0.5)
fig.show()
```

**3. Sub-Category Sales Analysis**

ðŸ”¸ Perform sales analysis at the sub-category level.

```python
sales_by_subcategory=df.groupby("Sub-Category")["Sales"].sum().reset_index()
sales_by_subcategory
```

```python
fig=px.bar(sales_by_subcategory,
           x="Sub-Category",
           y="Sales",
           color="Sub-Category",
           title="Sales Analysis by Sub-Category")
fig
```

**4. Monthly Profit Analysis**

ðŸ”¸ Determine which month generated the highest profit.

```python
Profit_by_month=df.groupby("order Month")["Profit"].sum().reset_index()
Profit_by_month
```

```python
month_order = list(month_label.values())
Profit_by_month["order Month"] = pd.Categorical(Profit_by_month["order Month"], categories=month_order, ordered=True)
Profit_by_month = Profit_by_month.sort_values("order Month")
Profit_by_month
```

```python
fig=px.line(Profit_by_month,
            x="order Month",
            y="Profit",
            title="Monthly_Profit_Analysis")
fig.show()
```

**Profit by Category**


ðŸ”¸ Analyze which categories brought in the most profit.

```python
profit_by_Category=df.groupby("Category")["Profit"].sum().reset_index()
profit_by_Category
```

```python
fig=px.pie(profit_by_Category,
           names="Category",
           values="Profit",
           color_discrete_sequence=px.colors.sequential.RdBu,
           hole=0.5,
           title="Profit by Category")
fig.update_layout(title_text="Profit by Category",title_font=dict(size=24),title_x=0.5)

fig.show()
```

**Profit By Sub Category**


ðŸ”¸ Analyze which sub-categories brought in the most profit.

```python
profit_by_sub_category=df.groupby("Sub-Category")["Profit"].sum().reset_index()
profit_by_sub_category
```

```python
fig=px.bar(profit_by_sub_category,
           x="Sub-Category",
           y="Profit",
           color="Sub-Category",
           title="Profit By Sub Category")
fig.show()
```

**Sales and Profit by Customer Segment**

ðŸ”¸ Study sales and profit across different customer segments.

```python
df.head()
```

```python
sales_profit_by_segment = df.groupby("Segment")[["Sales", "Profit"]].sum().reset_index()

fig = px.bar(sales_profit_by_segment,
             x="Segment",
             y=["Sales", "Profit"],
             title="Sales and Profit by Customer Segment",
             barmode='group')

fig.show()
```

**Sales to Profit Ratio**

ðŸ”¸ Calculate and analyze the sales-to-profit ratio for better efficiency evaluation.

```python
sales_profit_by_segment = df.groupby("Segment")[["Sales", "Profit"]].sum().reset_index()
sales_profit_by_segment
```

```python
sales_profit_by_segment["Sales to Profit Ratio"] = sales_profit_by_segment["Sales"] / sales_profit_by_segment["Profit"]
sales_profit_by_segment
```

```python
sales_profit_by_segment[["Segment","Sales to Profit Ratio"]]
```

**Sales and Profit by State**

```python
sales_profit_by_state = df.groupby("State")[["Sales", "Profit"]].sum().reset_index()

fig = px.bar(sales_profit_by_state,
             x="State",
             y=["Sales", "Profit"],  # Plotting both Sales and Profit
             title="Sales and Profit by State",
             barmode='group')
fig.update_layout(xaxis_tickangle=-45,title_font=dict(size=24),title_x=0.5)
fig.show()



```

```python
sales_profit_by_year = df.groupby("order Year")[["Sales", "Profit"]].sum().reset_index()

fig = px.line(sales_profit_by_year,
              x="order Year",
              y=["Sales", "Profit"], # Plotting both Sales and Profit
              title="Sales and Profit by Year")

fig.update_layout(
    xaxis = dict(
        tickmode = 'linear', # Ensure all years are shown on the x-axis
        dtick = 1           # Set tick interval to 1 to show every year
    )
)

fig.show()
```
